```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
# 1- Abre el fichero de `rating-and-performance` con el método que sea más correcto.
df_rating = pd.read_csv("files/rating-and-performance.csv" )
df_rating.head(2)
```

```python
# 2- Pongámos en práctica el `loc` e `iloc`:
# - Usando el `loc`: 
# - Extraed todas las filas de la columna "price". 

df_rating.loc[:, "price"]
```

```python
# Extraed todas las filas de las columnas "price" y "retail_price".
df_rating.loc[:, ["price", "retail_price"]]
```

```python
# Extraed las filas pares, de las columnas "rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count". 
df_rating.loc[::2, ["rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count"]]
```

```python
# Extraed las las filas cuyo precio sea mayor que 10 y sus unidades vendidas ("units_sold") sea mayor que 50. 

df_rating.loc[ (df_rating["price"] > 10) & (df_rating["units_sold"] > 5000)].head(1)
```

```python
# usando el iloc
# Extraed todas las filas de la columna "price". 
df_rating.iloc[:, 3]
```

```python
# Extraed todas las filas de las columnas "price" y "retail_price"
df_rating.iloc[:, 3:5]
```

```python
# Extraed las filas pares, de las columnas "rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count". 
df_rating.iloc[::2, 3:15]
```

```python
# Extraed las las filas cuyo precio sea mayor que 10 y sus unidades vendidas ("units_sold") sea mayor que 50. 

df_rating.iloc[ list((df_rating["price"] > 10) & (df_rating["units_sold"] > 5000))].head(1)
```
